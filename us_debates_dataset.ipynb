```python
import pyreadr
import pandas as pd
```

```python
data = pyreadr.read_r("data/debate_transcripts.rda")
```

```python
df = data["debate_transcripts"]
df = df[df["type"] == "Pres"]
```

```python
df
```

```python
years = df["election_year"].unique()
dates = df["date"].unique()
```

```python
dates
```

```python
first_debates = (
    df[["election_year", "date"]]
    .sort_values(by="date", ascending=False)
    .drop_duplicates(subset=["election_year"], keep="last")["date"]
    .values
)
first_debates
```

```python
df = df[df["date"].isin(first_debates)]
```

```python
df
```

```python
df.to_csv("data/first_debates.csv", index=False)
```

```python
for year in years:
    date = df[df["election_year"] == year]["date"].values[0]
    df[df["election_year"] == year].to_csv(f"data/by_date/{date}.csv", index=False)
```
